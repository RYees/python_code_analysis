```python
import matplotlib.pyplot as plt
import os, sys, json
rpath = os.path.abspath('..')
if rpath not in sys.path:
    sys.path.insert(0, rpath)

import chroma as utils
```

```python
import pandas as pd
def json_reader(file_path):
    with open(file_path, 'r') as file:
        json_data = json.load(file)
    return json_data

pdftexts = json_reader(file_path = './json.json')
df = pd.DataFrame(pdftexts)
```

```python
# Extract the 'user_id' and 'code_content' columns from the DataFrame
user_ids = df['user_id'].tolist()
code_contents = df['code_content'].tolist()

# Combine the 'code_content' values into a single string
combined_text = '\n\n'.join(code_contents)

# Call the text splitter function with the combined text
character_split_texts = utils.text_splitter_chunks([combined_text])

# Print the output
print(f"Total chunks: {len(character_split_texts)}")
```

```python
token_split_texts = utils.sentence_transfomer_textsplitter(character_split_texts)
print(token_split_texts[10])
print(f"\nTotal chunks: {len(token_split_texts)}")
```

```python
embedding_function = utils.embedding(token_split_texts)
print(embedding_function([token_split_texts[10]]))
```

```python
chroma_collection = utils.connect_with_chromadb(embedding_function, token_split_texts)
chroma_collection.count()
```

```python
query = "From the information that you have what is the commonly used python code in the document?"
retrieved_documents = utils.vectordb_answer_question(query, chroma_collection)
for document in retrieved_documents:
        print(document)
        print('\n')
```

```python
output_answer = utils.openai_model_answer(query=query, retrieved_documents=retrieved_documents)
print(output_answer)
```
